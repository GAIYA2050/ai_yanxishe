```python
import os
import numpy as np
import pandas as pd
```

```python
x_train = pd.read_csv('../data/train_demographic_info.csv',
                    names=['patient_id', 'age', 'sex', 'adult_bmi', 'child_weight', 'child_height'])
x_train.head()
```

```python
y_train = pd.read_csv('../data/train_patient_diagnosis.csv', names=['patient_id', 'diagnosis'])
y_train.head()
```

```python
train = pd.merge(x_train, y_train, how='inner', on='patient_id')
train.head()
```

```python
train.fillna('', inplace=True)
train.head()
```

```python
new_train = pd.DataFrame()
old_path = '../data/train'
new_path = '../data/input'
index = 0
file_prefix_set = set()
for file_name in os.listdir(old_path):
    patient_id = file_name.split('_')[0]
    file_prefix = file_name.split('.')[0]
    if file_prefix in file_prefix_set:
        continue
    else:
        file_prefix_set.add(file_prefix)
    new_path_dir = new_path + '/' + str(index)
    if not os.path.exists(new_path_dir):
        os.makedirs(new_path_dir)
#         print('{}/{}/{}'.format(old_path, dir_name, file_name))
#         print('{}/{}'.format(new_path_dir, file_name))
#     os.system("cp {}/{}.txt {}/{}.txt".format(old_path, file_prefix, new_path_dir, file_prefix))
#     os.system("cp {}/{}.wav {}/{}.wav".format(old_path, file_prefix, new_path_dir, file_prefix))
#         print(train[train['patient_id'] == int(patient_id)])
    new_train = new_train.append(train[train['patient_id'] == int(patient_id)], ignore_index=True)
    patient_id_index = list(new_train.columns).index('patient_id')
    new_train.iloc[index, patient_id_index] = index
#         print(new_train)
    index += 1
```

```python
new_train.tail()
```

```python
new_train['audio_and_txt_files_path'] = new_train['patient_id']
new_train.to_csv('../data/input/train.csv', index=None)
```

```python
test = pd.read_csv('../data/test_demographic_info.csv',
                    names=['patient_id', 'age', 'sex', 'adult_bmi', 'child_weight', 'child_height'])
test.head()
```

```python
test.fillna('', inplace=True)
test.head()
```

```python
origin_index = 769
index = origin_index
new_test = pd.DataFrame()
old_path = '../data/test'
new_path = '../data/input'
file_prefix_set = set()
ids = []
for file_name in os.listdir(old_path):
    patient_id = file_name.split('_')[0]
    file_prefix = file_name.split('.')[0]
    if file_prefix in file_prefix_set:
        continue
    else:
        file_prefix_set.add(file_prefix)
    new_path_dir = new_path + '/' + str(index)
    if not os.path.exists(new_path_dir):
        os.makedirs(new_path_dir)
#     os.system("cp {}/{}.txt {}/{}.txt".format(old_path, file_prefix, new_path_dir, file_prefix))
#     os.system("cp {}/{}.wav {}/{}.wav".format(old_path, file_prefix, new_path_dir, file_prefix))
    new_test = new_test.append(test[test['patient_id'] == int(patient_id)], ignore_index=True)
    patient_id_index = list(new_test.columns).index('patient_id')
    new_test.iloc[index-origin_index, patient_id_index] = index
    ids.append(patient_id)
    index += 1
```

```python
new_test['id'] = ids
new_test.head()
```

```python
new_test['audio_and_txt_files_path'] = new_test['patient_id']
new_test.to_csv('../data/input/test.csv', index=None)
```

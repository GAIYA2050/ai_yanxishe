```python
import numpy as np
import pandas as pd
```

```python
pred = pd.read_csv('../data/pred.csv', names=['patient_id', 'diagnosis'])
pred['patient_id'] = pred['patient_id']
pred.head()
```

```python
test = pd.read_csv('../data/input/test.csv')
test['patient_id']  = test['patient_id'] - 769
test.head()
```

```python
test = pd.merge(test, pred, how='inner', on='patient_id')
test.head()
```

```python
id_list = np.unique(test['id'].tolist())
id_list.sort()
diagnosis = []
for id in id_list:
    diagnosis_df = test[test['id'] == id]
    ans = dict()
    for j, vv in diagnosis_df.iterrows():
        if vv['diagnosis'] not in ans:
            ans[vv['diagnosis']] = 1
        else:
            ans[vv['diagnosis']] += 1
    ans = sorted(ans.items(), key=lambda x: x[1], reverse=True)
    print(ans)
    diagnosis.append(ans[0][0])
```

```python
submission = pd.DataFrame({'patient_id': id_list, 'diagnosis': diagnosis})
submission.to_csv('../data/submission.csv', index=None, header=None)
submission.head()
```
